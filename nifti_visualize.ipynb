```python
import matplotlib.pyplot as plt
import numpy as np
import nibabel as nib
# conda install nibabel -y or pip install nibabel로 nibabel 설치
```

```python
# 데이터 경로 - ~~~~_t1.nii.gz 파일 시각화 가능
# 우리가 볼 데이터는 t1 데이터입니다.
data_location = '/home/geonwoo/nas-data/brats2021/BraTS2021_00000/BraTS2021_00000_t1.nii.gz'
```

```python
# 1. 데이터 로드 (img가 nifti class, data는 ndarray 형태)
img = nib.load(data_location)
data = img.get_fdata()
data.shape # x,y,z
```

```python
img.header.get_zooms() # 해상도 확인, 1mm, 충분
```

```python
# 2. 데이터 시각화 (중간 슬라이스를 예시로 표시)
slice_0 = data[data.shape[0] // 2, :, :]  # Sagittal (좌우)
slice_1 = data[:, data.shape[1] // 2, :]  # Coronal (앞뒤)
slice_2 = data[:, :, data.shape[2] // 2]  # Axial (위아래)

# 3. Plotting
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(slice_0.T, cmap='gray', origin='lower')
axes[0].set_title("Sagittal Slice")
axes[1].imshow(slice_1.T, cmap='gray', origin='lower')
axes[1].set_title("Coronal Slice")
axes[2].imshow(slice_2.T, cmap='gray', origin='lower')
axes[2].set_title("Axial Slice")

for ax in axes:
    ax.axis("off")

plt.show()
```
